### Distribuição de Variáveis Quantitativas

````python
quantitativas = df.select_dtypes(include=['number']).columns

# Tamanho da figura
plt.figure(figsize=(16, 12))

# Loop para gerar histogramas e gráficos de densidade
for i, col in enumerate(quantitativas, 1):
    plt.subplot(3, 3, i)  # Ajuste o layout conforme necessário
    sns.histplot(df[col], kde=True, bins=30)  # Histogram + KDE
    plt.title(f'Distribuição de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')

# Evitar sobreposição
plt.tight_layout()
plt.show()
````

A análise da distribuição das variáveis quantitativas, como renda, revela uma maioria com distribuição assimétrica e picos em faixas específicas. A presença de outliers sugere discrepâncias significativas entre grupos, enquanto variáveis como posse de imóvel e escolaridade mostram concentração em categorias determinadas. Os gráficos de densidade ajudam a identificar tendências gerais, e compreender essas distribuições é fundamental para decisões e análises estatísticas mais robustas.

### Frequência de Variáveis Qualitativas

````python
plt.rc('figure', figsize=(12,6))

for col in qualitativas:
    ax = sns.countplot(x='data_ref', hue=col, data=df)
    tick_labels = df['data_ref'].map(lambda x: x.strftime('%b/%Y')).unique()

    ax.set_xticklabels(labels=tick_labels, rotation=45)
    ax.legend(bbox_to_anchor=(1,.5), loc=6, title=f"'{col}'")

    plt.show()
````

A visualização da frequência das variáveis qualitativas ao longo do tempo revela tendências e mudanças nas categorias, com picos e flutuações que indicam comportamentos variáveis entre os grupos. Os gráficos de contagem mostram que a posse de imóvel tem alta frequência, sugerindo uma correlação com segurança financeira. Escolaridade e estado civil também são significativos, refletindo tendências sociais e econômicas. As mudanças nessas variáveis estão associadas a fatores externos, como políticas econômicas, influenciando a dinâmica da renda e preferências sociais.

### Tendências e Distribuições: Relação entre Variáveis Qualitativas e Renda ao Longo do Tempo

````python
plt.rc('figure', figsize=(16,4))

for col in qualitativas:
    fig, axes = plt.subplots(1,2)
    fig.subplots_adjust(wspace=.6)
    
    tick_labels = df['data_ref'].map(lambda x: x.strftime('%b/%Y')).unique()
    
    # barras empilhadas:
    df_crosstab = pd.crosstab(index=df['data_ref'], 
                              columns=df[col], 
                              normalize='index')
    ax0 = df_crosstab.plot.bar(stacked=True, ax=axes[0])
    ax0.set_xticklabels(labels=tick_labels, rotation=45)
    axes[0].legend(bbox_to_anchor=(1,.5), loc=6, title=f"'{col}'")

    # perfis médios no tempo:
    ax1 = sns.pointplot(x='data_ref', y='renda', hue=col, data=df, dodge=True, ci=95, ax=axes[1])
    ax1.set_xticklabels(labels=tick_labels, rotation=45)
    axes[1].legend(bbox_to_anchor=(1,.5), loc=6, title=f"'{col}'")

    plt.show()
````

A análise revela que a posse de imóvel é uma variável predominante, correlacionada com a estabilidade financeira ao longo do tempo. A visualização das barras empilhadas mostra variações nas categorias qualitativas, indicando mudanças nas preferências sociais. Os perfis médios da renda revelam flutuações associadas a diferentes níveis de escolaridade e estado civil, refletindo tendências sociais e econômicas. Essas dinâmicas temporais sugerem que fatores externos, como políticas econômicas, podem influenciar as escolhas individuais, impactando a renda e a segurança financeira ao longo do tempo.
