1) Avalie a estrutura de correlação das variáveis quantitativas através de um gráfico de "matriz de dispersão" e através da avaliação gráfica da matriz de correlações. Veja se você identifica algum padrão interessante ou que te faça sentido.

````python
df.drop(columns=['Unnamed: 0', 'index', 'mau'], inplace=True)

print(df.info(), '\n')
````

````python
print('Quantidade de linhas duplicadas:', df.duplicated().sum())
````

````python
# Matriz de correlação

df.corr()
````

````python
# Matriz de dispersão

sns.pairplot(data=df, 
             hue='tipo_renda', 
             vars=['qtd_filhos', 
                   'idade', 
                   'tempo_emprego', 
                   'qt_pessoas_residencia', 
                   'renda'])

plt.show()
````

---

 >A análise preliminar mostra que a renda tende a aumentar com a idade e tempo de emprego, variando significativamente entre tipos de renda. Fatores como número de filhos e pessoas na residência também influenciam, mas de forma menos direta. Para aprofundar, sugere-se quantificar correlações, comparar grupos e visualizar distribuições.
 ---

````python
# Renomeando as colunas
df_renomeado = df.rename(columns={
    'qtd_filhos': 'Número de Filhos',
    'idade': 'Idade',
    'tempo_emprego': 'Tempo de Emprego',
    'qt_pessoas_residencia': 'Nº de Pessoas na Residência',
    'renda': 'Renda'
})

# Calculando a correlação e plotando o heatmap com os novos nomes
corr = df_renomeado[['Número de Filhos', 'Idade', 'Tempo de Emprego', 'Nº de Pessoas na Residência', 'Renda']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
````

---
>Interessante desses dados é a forte correlação positiva (0.9) entre o “Número de Filhos” e o “Nº de Pessoas na Residência”. Isso sugere que, conforme o número de filhos aumenta, o número total de pessoas na residência também tende a aumentar significativamente. Esse padrão pode ser útil para políticas habitacionais e planejamento urbano, indicando a necessidade de moradias maiores para famílias com mais filhos.

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
df_log = df.copy()
df_log['renda'] = np.log(df_log['renda'])

plt.figure(figsize=(16,9))

sns.scatterplot(x='tempo_emprego',  # <- 1ª
                y='renda', 
                hue='tipo_renda', 
                size='idade',  # <- 2ª
                data=df_log,
                alpha=0.4)

# Linha de tendência:
sns.regplot(x='tempo_emprego', 
            y='renda', 
            data=df_log, 
            scatter=False, 
            color='.3')

plt.show()
````

````python
df_log = df.copy()
df_log['renda'] = np.log(df_log['renda'])

plt.figure(figsize=(16,9))

sns.scatterplot(x='idade',  # <- 1ª
                y='renda', 
                hue='tipo_renda', 
                size='tempo_emprego',  # <- 2ª
                data=df_log,
                alpha=0.4)

# Linha de tendência:
sns.regplot(x='idade', 
            y='renda', 
            data=df_log, 
            scatter=False, 
            color='.3')

plt.show()
````

5) A aplicação do LOG você acha que melhorou ou piorou a análise?

````python
# Diferença após aplicação do LOG:

pd.concat(objs=[df.corr().tail(1), 
                df_log.corr().tail(1)], 
          axis=0
         ).reset_index(drop=True).rename(index={0:'renda', 
                                                1:'renda_log'})
````

---
>A matriz de correlação mostra que a aplicação do logaritmo alterou as correlações entre as variáveis, com algumas aumentando e outras diminuindo. Visualmente, a análise melhorou, especialmente na correlação entre renda e tempo de emprego, pois houve uma redução na dispersão dos pontos, o que os concentrou e facilitou a compreensão da linha de tendência, que aparenta uma inclinação maior. Quanto à variável idade, apesar de a inclinação permanecer a mesma, a relação entre os valores inferiores e superiores ficou mais clara.
