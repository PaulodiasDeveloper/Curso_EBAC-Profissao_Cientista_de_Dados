1. Considere o script que monta a análise bivariada que você fez na tarefa anterior. Transforme esse script em uma função, que deve:
- Ter como parâmetros de entrada:
    - Um *dataframe* contendo os dados a serem avaliados
    - Um *string* contendo o nome da variável resposta
    - Um *string* contendo o nome da variável explicativa
- E deve retornar um *dataframe* com os dados da bivariada. 
**Monte** a mesma bivariada pelo menos três variáveis qualitativas do *data-frame*. Qual delas parece discriminar mais o risco?

````python
def bivariada_qualitativa(dataframe:pd.DataFrame, x:str, y:str) -> pd.DataFrame:
    
    tabela_cruzada = pd.crosstab(index=dataframe[x], 
                                 columns=dataframe[y], 
                                 margins=True, 
                                 margins_name='total')
    
    tabela_cruzada['prob'] = tabela_cruzada[1] / tabela_cruzada['total']
    
    tabela_cruzada['odds'] = tabela_cruzada[1] / tabela_cruzada[0]
    tabela_cruzada['odds_ratio'] = tabela_cruzada['odds'] / tabela_cruzada.loc['total', 'odds']

    tabela_cruzada['logito'] = np.log(tabela_cruzada['odds'])
    tabela_cruzada['WOE'] = np.log(tabela_cruzada['odds'] / tabela_cruzada.loc['total', 'odds'])
    
    return tabela_cruzada
````

2. Monte uma função semelhante para categorizar variáveis quantitativas contínuas (com muitas categorias) como ```age```.  
    Além dos mesmos parâmetros da função anterior, defina mais um parâmetro como número de categorias que você deseja quebrar. Defina um valor '*default*' de 5 grupos para este parâmetro.  

````python
def bivariada_quantitativa(dataframe: pd.DataFrame, 
                           x: str, 
                           y: str, 
                           q: int = 5
                          ) -> pd.DataFrame:
    # Garantir que a coluna x é numérica
    dataframe[x] = pd.to_numeric(dataframe[x], errors='coerce')
    
    # Tratar valores ausentes
    dataframe[x].fillna(dataframe[x].mean(), inplace=True)
    
    # Criar tabela cruzada com cortes
    tabela_cruzada = (pd.crosstab(index=dataframe[x], 
                                  columns=dataframe[y], 
                                  margins=True, 
                                  margins_name='total')
                        .groupby(by=pd.qcut(x=dataframe[x], 
                                            q=q, 
                                            duplicates='drop'))
                        .agg(func='sum'))
    
    # Adicionar total
    tabela_cruzada = pd.concat(objs=[tabela_cruzada, 
                                     pd.DataFrame(data=tabela_cruzada.sum(), 
                                                  columns=['total']).T])
    tabela_cruzada.index.name = x
    
    # Calcular probabilidade, odds e razão de chances
    tabela_cruzada['prob'] = tabela_cruzada[1] / tabela_cruzada['total']
    tabela_cruzada['odds'] = tabela_cruzada[1] / tabela_cruzada[0]
    tabela_cruzada['odds_ratio'] = tabela_cruzada['odds'] / tabela_cruzada.loc['total', 'odds']

    # Calcular logito e WOE
    tabela_cruzada['logito'] = np.log(tabela_cruzada['odds'])
    tabela_cruzada['WOE'] = np.log(tabela_cruzada['odds'] / tabela_cruzada.loc['total', 'odds'])
    
    # Calcular média dos valores por intervalo
    tabela_cruzada['media_quantitativa'] = (
        dataframe.groupby(pd.qcut(dataframe[x], q=q, duplicates='drop'))[x]
        .mean()
    )
    tabela_cruzada.loc['total', 'media_quantitativa'] = dataframe[x].mean()
    tabela_cruzada['media_quantitativa'] = round(tabela_cruzada['media_quantitativa'])
    
    return tabela_cruzada
````

````python
bivariada_quantitativa(dataframe=df, x='age', y='flag_doente')
````

3. Construa um modelo de regressão logística com as variáveis qualitativas: ```sex + cp +  trestbps``` e com a variável quantitativa ```age```.

**Interprete os parâmetros.**

````python
reglog = smf.logit(formula='''flag_doente ~ 
                              sex 
                              + C(cp, Treatment(reference=4))
                              + trestbps 
                              + age
                              ''', data=df).fit()
reglog.summary()
````

**Interpretação Aprimorada dos Parâmetros:**

* **coef (Coeficiente):**
    * Este valor representa a magnitude da influência de cada variável independente (sexo, tipo de dor no peito, pressão arterial em repouso, idade) na probabilidade da presença de doença cardíaca.
    * Um coeficiente positivo indica que um aumento na variável independente está associado a um aumento na probabilidade de doença cardíaca, enquanto um coeficiente negativo indica o contrário.
    * O coeficiente de "sexo" sendo o mais alto sugere que o sexo do indivíduo tem um impacto significativo na probabilidade de doença cardíaca, em relação as outras variaveis analisadas.
* **std err (Erro Padrão):**
    * O erro padrão mede a precisão da estimativa do coeficiente. Valores menores indicam maior precisão.
    * Um erro padrão baixo em relação ao coeficiente sugere que a estimativa é confiável e que a variável independente tem um efeito real na variável dependente.
* **z (Estatística Z):**
    * A estatística Z é a razão entre o coeficiente e seu erro padrão.
    * Ela indica quantas unidades de erro padrão o coeficiente se desvia de zero.
    * Valores absolutos altos da estatística Z indicam que a variável independente é estatisticamente significativa.
    * O cp (tipo de dor no peito) sendo o valor absoluto mais alto, demonstra que essa variavel tem grande poder de explicação sobre a doença cardiaca.
* **P>|z| (Valor P):**
    * O valor P representa a probabilidade de observar um efeito tão extremo quanto o observado, assumindo que não há efeito real.
    * Valores P menores que 0,05 são geralmente considerados estatisticamente significativos, indicando que é improvável que o efeito observado seja devido ao acaso.
    * Todos os valores p analisados estão abaixo de 0,05, o que reforça a afirmação de que todas as variaveis independentes analisadas são estatisticamente significantes para explicar a doença cardiaca.

**Conclusões Refinadas:**

* Todas as variáveis independentes analisadas (sexo, tipo de dor no peito, pressão arterial em repouso, idade) demonstraram significância estatística na previsão da presença de doença cardíaca, com base nos baixos valores P.
* O sexo do indivíduo parece ser o fator de maior influência, seguido pelo tipo de dor no peito, demonstrado pelo valor absoluto da estatistica z.
* A pressão arterial em repouso e a idade também contribuem significativamente para o modelo, embora em menor grau.
* A precisão dos coeficientes, indicada pelos baixos erros padrão, reforça a confiabilidade dessas conclusões.

* É importante lembrar que correlação não implica causalidade, e que esses resultados indicam associações estatísticas, não relações de causa e efeito diretas.

Ao incorporar essas melhorias, a interpretação dos parâmetros se torna mais completa e fornece uma visão mais profunda da relação entre as variáveis independentes e a doença cardíaca.

4. Avalie o seu modelo quanto a **calibragem**:
- Calcule a probabilidade de evento predita segundo o seu modelo
- Categorize essa probabilidade em G=5 grupos
- Calcule a probabilidade de evento predita média por grupo
- Calcule a taxa de eventos (média da variável indicadora de eventos) por grupo
- Compare graficamente o valor eperado versus observado para a taxa de maus por grupo

````python
# Calculando a probabilidade de evento predita segundo o modelo:

df['predito'] = reglog.predict(exog=df)
df
````

````python
# Categorizando essa probabilidade em 5 grupos:

pred_cat = pd.qcut(x=df['predito'], q=5)
# pred_cat.value_counts().sort_index()

group_reg = df.groupby(by=pred_cat)

qualid = (group_reg[['flag_doente']].count().rename(columns={'flag_doente':'contagem'}))
qualid
````

````python
# Calculando a probabilidade de evento predita média por grupo:

qualid['predito'] = group_reg['predito'].mean()
qualid
````

````python
# Calculando a taxa de eventos (média da variável indicadora de eventos) por grupo:

qualid['pct_doente'] = group_reg['flag_doente'].mean()
qualid
````

````python
# Comparando graficamente o valor esperado versus observado para a taxa de maus por grupo:

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax = qualid['pct_doente'].plot(label='% Predito')
ax = qualid['predito'].plot(label='% Observado')

ticks = ax.set_xticks(ticks=[0,1,2,3,4])
labels = ax.set_xticklabels([1,2,3,4,5])
ax.legend()
ax.set_xlabel(xlabel='Grupo')
ax.set_ylabel(ylabel='Probabilidade de evento')

plt.show()
````

5. Avalie o seu modelo quanto a discriminação calculando acurácia, GINI e KS.

````python
# Acurácia
acc = metrics.accuracy_score(df['flag_doente'], df['predito'] > .5)
print('Acurácia: {0:.2f}%'.format(acc*100))

# AUC
fpr, tpr, thresholds = metrics.roc_curve(df['flag_doente'], df['predito'])
auc_ = metrics.auc(fpr, tpr)

# GINI
gini = 2 * auc_ -1

# KS
ks = ks_2samp(df.loc[df['flag_doente'] == 1, 'predito'], df.loc[df['flag_doente'] != 1, 'predito']).statistic

# print(f'AUC: {auc_*100:.2f}%')
print(f'GINI: {gini*100:.2f}%')
print(f'KS: {ks*100:.2f}%')
````
