### Alternativa usando o Usando **TvDatafeed**

```python
# Inicializa a conexão (você pode incluir usuário/senha se necessário)
tv = TvDatafeed()
```

```python
ativos = [
    ('BTCUSDT', 'BINANCE'),
    ('ETHUSDT', 'BINANCE'),
    ('USDCUSDT', 'BINANCE')
]
```

```python
dados_ativos = {}

for symbol, exchange in ativos:
    try:
        print(f'Buscando dados de {symbol} na {exchange}...')
        dados = tv.get_hist(symbol=symbol, exchange=exchange, interval=Interval.in_daily, n_bars=1000)
        dados_ativos[symbol] = dados
    except Exception as e:
        print(f'Erro ao buscar {symbol}: {e}')
```

```python
# Salvar os dados em um arquivo CSV
for symbol, dados in dados_ativos.items():
    dados.to_csv(f'{symbol}.csv')
```

```python
# Carregar os dados de um arquivo CSV
df_btc = pd.read_csv('BTCUSDT.csv', index_col=0)
df_eth = pd.read_csv('ETHUSDT.csv', index_col=0)
df_usdc = pd.read_csv('USDCUSDT.csv', index_col=0)

# Concatenar os DataFrames
df_all = pd.concat([df_btc, df_eth, df_usdc])

df_all.head()
```

```python
# Salvar os dados em um arquivo CSV
df_all.to_csv('ativos.csv')
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
df_all['amplitude_diaria'] = df_all['high'] - df_all['low']
```

```python
df_all.head()
```

2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.

```python
df_all = df_all.reset_index()
```

```python
# Criar gráfico
fig = px.line(df_all, 
              x='datetime', 
              y='amplitude_diaria', 
              color='ativo',
              title='Variação da Amplitude Diária por Ativo',
              labels={
                  'datetime': 'Data',
                  'amplitude_diaria': 'Amplitude Diária (High - Low)',
                  'ativo': 'Ativo'
              })

fig.update_layout(xaxis_title='Data',
                  yaxis_title='Amplitude de Variação',
                  template='plotly_white')

fig.show()
```

3. Veja se consegue identificar que ativo apresentou maior volatilidade.
