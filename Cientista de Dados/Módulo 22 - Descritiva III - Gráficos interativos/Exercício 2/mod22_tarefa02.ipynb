#### 1. Carregue os dados da tarefa anterior

````python
import pandas as pd

import plotly.express as px
import matplotlib.pyplot as plt
from plotly import graph_objects
from plotly.graph_objs import Layout
import plotly.graph_objects as go

from tvDatafeed import TvDatafeed, Interval
````

````python
df = pd.read_csv(filepath_or_buffer='ativos.csv', index_col=0)
df
````

````python
df['symbol'] = df['symbol'].str.replace('BINANCE:', '')

print(df['symbol'].unique())
````

````python
df = df.rename(columns={'symbol': 'ativo'})

df
````

````python
df_wide = df.reset_index().pivot(index='datetime', columns='ativo', values='close')
df_wide.columns.name = None  # remove o nome 'ativo' do eixo das colunas

print(df_wide)
````

#### 2. Média móvel





1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu

````python
df_ma = df_wide.rolling(window=90).mean()
df_ma
````

2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima

````python
df_std = df_wide.rolling(window=90).std()
df_std
````

3. Monte um gráfico de linhas para as médias móveis dos ativos da sua base

````python
df_ma = df_ma.stack().reset_index()
df_ma.columns = ['data', 'ativo', 'valor']

fig = px.line(data_frame=df_ma, 
              x='data', 
              y='valor', 
              color='ativo', 
              template='plotly_white')
fig.show()
````

4. Monte um gráfico de linhas para os desvios padrão em janela móvel

````python
df_std = df_std.stack().reset_index()
df_std.columns = ['data', 'ativo', 'valor']

fig = px.line(data_frame=df_std, 
              x='data', 
              y='valor', 
              color='ativo', 
              template='plotly_white')
fig.show()
````

#### 3. Gráfico de *candlestick*

1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base

````python
print(df_wide.columns)
````

````python
df_last60 = df_wide[['BTCUSDT', 'ETHUSDT', 'USDCUSDT']].tail(60)
df_last60 = df_last60.reset_index()

print(f'Quantidade de linhas do período selecionado: {df_last60.shape[0]}.\n')
df_last60
````

2. Monte um gráfico de *candlestick* para esses dados.

````python

df_last60['Close'] = df_last60['BTCUSDT']
df_last60['Open'] = df_last60['BTCUSDT'].shift(1)
df_last60['Open'] = df_last60['Open'].fillna(df_last60['Close'])

df_last60['High'] = df_last60[['Open', 'Close']].max(axis=1) * 1.01
df_last60['Low'] = df_last60[['Open', 'Close']].min(axis=1) * 0.99


fig = go.Figure(data=[go.Candlestick(
    x=df_last60['datetime'],
    open=df_last60['Open'],
    high=df_last60['High'],
    low=df_last60['Low'],
    close=df_last60['Close'],
    name='BTCUSDT'
)])

fig.update_layout(
    title='índice Bitcoin (BTCUSDT) simulado',
    xaxis_title='Data',
    yaxis_title='Preço BTCUSDT',
    xaxis_rangeslider_visible=False
)

fig.show()
````
