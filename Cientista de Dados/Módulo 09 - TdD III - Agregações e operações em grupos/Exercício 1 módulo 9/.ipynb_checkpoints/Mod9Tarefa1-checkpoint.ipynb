### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
# Filtrar as colunas relevantes
cols = ['munResNome', 'IDADEMAE', 'IDADEPAI']

# Filtrar as linhas onde a idade dos pais não é nula
filtered_df = sinasc_raw[cols].dropna(subset=['IDADEPAI'])

# Calcular a média das idades das mães e dos pais por município
idade_media = filtered_df.groupby('munResNome').agg({
    'IDADEMAE': 'mean',
    'IDADEPAI': 'mean'
}).reset_index()

# Renomear as colunas para clareza
idade_media.columns = ['Municipio', 'Idade Media Mae', 'Idade Media Pai']

print(idade_media)
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc_1 = sinasc_raw[['ESCMAE', 'IDADEMAE', 'DTNASC']]
sinasc_1.head()
```

```python
sinasc_aniversario = '10-01'  # Formato 'mês-dia'

# Converter a coluna DTNASC para datetime
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])

# Fltrando os dados para incluir apenas as datas de aniversário específicas
filtros_data = sinasc_raw[sinasc_raw['DTNASC'].dt.strftime('%m-%d') == sinasc_aniversario]

# Exiba os dados filtrados
print(filtros_data)

# Agrupar por faixa de escolaridade da mãe e sexo, e calcular o peso médio
if not filtros_data.empty:
    resultado = filtros_data.groupby(['ESCMAE', 'SEXO'])['PESO'].mean().reset_index()
    print(resultado)
else:
    print("Nenhum bebê nasceu no dia especificado.")
```

```python
# Agrupando por faixa de escolaridade da mãe e sexo, e calculando o peso médio
if not filtros_data.empty:
    resultado = filtros_data.groupby(['ESCMAE', 'SEXO'])['PESO'].mean().reset_index()
    
    # Criando o gráfico de barras
    fig, ax = plt.subplots(figsize=(10, 6))
    resultado.pivot(index='ESCMAE', columns='SEXO', values='PESO').plot(kind='bar', ax=ax)
    
    # Configurações do gráfico
    ax.set_xlabel('Escolaridade da Mãe')
    ax.set_ylabel('Peso Médio')
    ax.set_title('Peso Médio de Bebês Nascidos em {}'.format(sinasc_aniversario))
    ax.legend(title='Sexo')
    
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
else:
    print("Nenhum bebê nasceu no dia especificado.")
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
print('Municípios em que nasceram menos bebês:\n')
print(sinasc_raw.munResNome.value_counts().sort_values().head())

sinasc_raw['COUNT'] = 1
sinasc_3 = sinasc_raw.groupby(by='munResNome')[['COUNT', 
                                                'IDADEMAE',
                                                'IDADEPAI']].agg(
    {'COUNT':[('Quantidade de bebês nascidos','count')], 
     'IDADEMAE':[('idade média','mean'), 
                 ('idade máxima',max), 
                 ('idade mínima',min)], 
     'IDADEPAI':[('idade média','mean'), 
                 ('idade máxima',max), 
                 ('idade mínima',min)]})

sinasc_3.sort_values(('COUNT', 'Quantidade de bebês nascidos'))
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_mar = sinasc_raw[sinasc_raw.DTNASC.dt.month == 3].copy()

print('O município que mais nasceu bebê no mês de março foi {} com {} nascimentos.'.format(
    sinasc_mar['munResNome'].value_counts().index[0], 
    sinasc_mar['munResNome'].value_counts().values[0]))

sinasc_mar['MES'] = sinasc_mar['DTNASC'].dt.month_name()
sinasc_mar['COUNT'] = 1
sinasc_4 = sinasc_mar.groupby(by=['MES',
                                  'munResNome']
                             )[['COUNT', 
                                'QTDFILVIVO',
                                'IDADEMAE',
                                'IDADEPAI']
                              ].agg({'COUNT':[('Qtd de bebês nascidos','count')], 
                                     'QTDFILVIVO':[('Média','mean'), 
                                                   ('Máx',max), 
                                                   ('Mín',min)], 
                                     'IDADEMAE':[('Média','mean'), 
                                                 ('Máx',max), 
                                                 ('Mín',min), 
                                                 ('Amplitude',lambda x: x.max()-x.min())], 
                                     'IDADEPAI':[('Média','mean'), 
                                                 ('Máx',max), 
                                                 ('Mín',min), 
                                                 ('Amplitude',lambda x: x.max()-x.min())]
                                    })

sinasc_4['Diferença da média IDADEPAI-IDADEMAE'] = sinasc_4[('IDADEPAI','Média')]-sinasc_4[('IDADEMAE','Média')]
sinasc_4.sort_values(('COUNT', 'Qtd de bebês nascidos'), 
                     ascending=False
                    )
```

```python
print('Mediana de todos os municípios no mês de março:')
round(sinasc_4.groupby(by='MES').median())
```

```python
print('Média de todos os municípios no mês de março:')
round(sinasc_4.groupby(by='MES').mean())
```
